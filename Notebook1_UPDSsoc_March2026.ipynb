Multiples of 3 or 5:


If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.
Find the sum of all the multiples of 3 or 5 below 1000.

```python
# We need a code that determines divisibility for 3 and 5; for that we use the modulo operator.
# We need a code that scans natrual numbers from 0 to 999; for that we use a for loop.
# Lets us try an if-else syntax for filtering what we need.
# We are asked for the sum, hence we add.

# Let us put everything in an empty list named Liszt
liszt = []

for i in range(1000):
  if i%3 == 0 or i%5 == 0:
    liszt.append(i)

print(sum(liszt))
```

- `liszt = []` creates an empty list to store the multiples.
- `range(1000)` generates numbers from 0 to 999, which are all natural numbers below 1000.
- `i % 3 == 0` checks if the number is divisible by 3.
- `i % 5 == 0` checks if the number is divisible by 5.
- `or` combines the two conditions so that numbers divisible by either 3 or 5 are included.
- `liszt.append(i)` adds the number to the list when the condition is true.
- `sum(liszt)` calculates the total sum of all multiples in the list.
- `print()` displays the final sum, which is 233168.
